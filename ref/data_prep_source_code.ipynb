### Disable SSL Checks

```python
import requests
import urllib3

def disable_ssl_verification():
    """Disable SSL certificate verification globally for requests."""
    urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

    original_request = requests.Session.request

    def request_without_ssl(self, *args, **kwargs):
        kwargs["verify"] = False
        return original_request(self, *args, **kwargs)

    requests.Session.request = request_without_ssl
```

### Now we have to do same for all the url and put it to the data frame

```python
all_link = df['Link'].tolist()
```

```python
import numpy as np

actual_content = []

for i, link in enumerate(all_link):
    print(f"Scraping content for link {i+1}/{len(all_link)}")

    response = requests.get(link)
    soup = BeautifulSoup(response.text, "html.parser")

    text_section = soup.find("div", class_="post-content-wrap")

    if text_section is None:
        actual_content.append(np.nan)
        continue

    post_text = ""

    for p in text_section.find_all("p"):
        post_text += p.get_text(strip=True) + " "

    actual_content.append(post_text.strip())
```

```python
from concurrent.futures import ThreadPoolExecutor
import numpy as np
def scrape_article(args):
    i, url = args
    print(f"Scraping {i+1}/{len(all_link)}")

    try:
        response = requests.get(url, timeout=10)

        if response.status_code != 200:
            return np.nan

        soup = BeautifulSoup(response.text, "html.parser")

        text_section = soup.find("div", class_="post-content-wrap")

        if text_section is None:
            return np.nan

        post_text = ""

        for p in text_section.find_all("p"):
            post_text += p.get_text(strip=True) + " "

        return post_text.strip()

    except Exception:
        return np.nan


with ThreadPoolExecutor(max_workers=100) as executor:
    actual_content = list(executor.map(scrape_article, enumerate(all_link)))
```

```python
actual_content
```

```python
df["Content"] = actual_content
df.head()
```
