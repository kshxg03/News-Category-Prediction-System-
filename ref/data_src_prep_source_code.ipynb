### Disable SSL Checks

````python
import requests
import urllib3

def disable_ssl_verification():
    """Disable SSL certificate verification globally for requests."""
    urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

    original_request = requests.Session.request

    def request_without_ssl(self, *args, **kwargs):
        kwargs["verify"] = False
        return original_request(self, *args, **kwargs)

    requests.Session.request = request_without_ssl
````

### Pattern to Navigate to Different Pages and Retrive Data

Most websites use one of the following pagination patterns:

#### 1. Page Number in the URL

```text
https://example.com/page/1/
https://example.com/page/2/
https://example.com/page/3/
```
`

````python
target_page = 10
all_post_data = {}  
for i in range(1, target_page + 1):
    print(f"Scraping  page {i}/{target_page}")
    if i == 1:
        page_url = f"{links['Politics']}"
    else:
        page_url = f"{links['Politics']}/page/{i}/"
    
    # Fetch the HTML content of the page
    page_response = requests.get(page_url)
    page_html_doc = page_response.text
    
    # Parse the HTML content using BeautifulSoup
    soup_page = BeautifulSoup(page_html_doc, 'html.parser')
    posts_section = soup_page.find("div", class_="listical-news-big")
    posts = posts_section.find_all("div", class_="ok-post-contents")
    
    # Extract post data from the current page
    for post in posts:
        for i in post.find_all("a"):
            all_post_data[i.string] = i.get("href")
            


all_post_data
````

````python
Target_Pages =1
Total_Data = {}

for category, url in links.items():
    for page in range(1, Target_Pages + 1):
        print(f"Scraping {category} page {page}/{Target_Pages}")

        if page == 1:
            page_url = url
        else:
            page_url = f"{url}/page/{page}/"

        # Fetch the HTML content
        page_response = requests.get(page_url)

        # Stop if the page doesn't exist
        if page_response.status_code != 200:    # this is how we handle page not exists situation
            break

        page_html_doc = page_response.text

        # Parse the HTML
        soup_page = BeautifulSoup(page_html_doc, "html.parser")
        posts_section = soup_page.find("div", class_="listical-news-big")

        if posts_section is None:    # if page found but the page dont have any available posts 
            break

        posts = posts_section.find_all("div", class_="ok-post-contents")

        # Extract post data
        for post in posts:
            for a in post.find_all("a"):
                Total_Data[a.string] = {"url":a.get("href"), "category": category}

    print()
````

````python
import requests
from bs4 import BeautifulSoup
from concurrent.futures import ThreadPoolExecutor, as_completed

Target_Pages = 700
Total_Data = {}

def scrape_page(category, url, page):
    print(f"Scraping {category} page {page}/{Target_Pages}")

    if page == 1:
        page_url = url
    else:
        page_url = f"{url}/page/{page}/"

    try:
        response = requests.get(page_url, timeout=10)

        if response.status_code != 200:
            return None

        soup = BeautifulSoup(response.text, "html.parser")
        posts_section = soup.find("div", class_="listical-news-big")

        if posts_section is None:
            return None

        data = {}

        posts = posts_section.find_all("div", class_="ok-post-contents")

        for post in posts:
            for a in post.find_all("a"):
                if a.string:
                    data[a.string] = {
                        "url": a.get("href"),
                        "category": category
                    }

        return data

    except Exception:
        return None


with ThreadPoolExecutor(max_workers=20) as executor:
    futures = []

    for category, url in links.items():
        for page in range(1, Target_Pages + 1):
            futures.append(executor.submit(scrape_page, category, url, page))

    for future in as_completed(futures):
        result = future.result()
        if result:
            Total_Data.update(result)

print(f"Total articles: {len(Total_Data)}")
````
