#### Getting the Website Page

```python
import requests

url = "https://english.onlinekhabar.com/"
response = requests.get(url)
print(response.status_code)
```

```python
html_doc = response.text
html_doc
```

#### Converting the html content into beautiful soup

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html_doc, 'html.parser')
print(soup.prettify())
```

```python
nav_bar=soup.find("ul", id="primary-menu")
nav_bar
```

```python
urls = nav_bar.find_all("a")
urls
```

```python
print(urls[0])
print(urls[0].get("href"))
print(urls[0].string)
```

```python
links={}
for url in urls:
    links[url.string] = url.get("href")

links
```

#### Collecting information

```python
# for politics page
politics_page = requests.get(links["Politics"])
politics_html_doc = politics_page.text
```

```python
# converting html into soup
soup_politics = BeautifulSoup(politics_html_doc, 'html.parser')
print(soup_politics.prettify())
```

```python
Posts_section = soup_politics.find("div", class_="listical-news-big")
Posts_section
```

```python
posts = Posts_section.find_all("div", class_="ok-post-contents")
posts
```

```python
post_data={}
for post in posts:
    for i in post.find_all("a"):
        post_data[i.string] = i.get("href")
        
post_data
```

```python
import requests
from bs4 import BeautifulSoup
from concurrent.futures import ThreadPoolExecutor, as_completed

Target_Pages = 7000
Total_Data = {}

def scrape_page(category, url, page):
    print(f"Scraping {category} page {page}/{Target_Pages}")

    if page == 1:
        page_url = url
    else:
        page_url = f"{url}/page/{page}/"

    try:
        response = requests.get(page_url, timeout=10)

        if response.status_code != 200:
            return None

        soup = BeautifulSoup(response.text, "html.parser")
        posts_section = soup.find("div", class_="listical-news-big")

        if posts_section is None:
            return None

        data = {}

        posts = posts_section.find_all("div", class_="ok-post-contents")

        for post in posts:
            for a in post.find_all("a"):
                if a.string:
                    data[a.string] = {
                        "url": a.get("href"),
                        "category": category
                    }

        return data

    except Exception:
        return None


with ThreadPoolExecutor(max_workers=20) as executor:
    futures = []

    for category, url in links.items():
        for page in range(1, Target_Pages + 1):
            futures.append(executor.submit(scrape_page, category, url, page))

    for future in as_completed(futures):
        result = future.result()
        if result:
            Total_Data.update(result)

print(f"Total articles: {len(Total_Data)}")
```
